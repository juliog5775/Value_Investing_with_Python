# VALUE INVESTING
Es una estrategía de inversion que se centra en encontrar compañias con un precio inferior a su valor intrinseco. Compañias con potencial de crecimiento en el largo plazo. 
Warren Buffett es un exponente de este tipo de inversion


https://github.com/juliog5775

```python
import yfinance as yf #pip install yfinance (si no esta instalada)
import pandas as pd
import numpy as np 
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
```

```python
table = pd.read_html(url) #Estraemos una tabla con todos los tickers del sp500 de Wikipedia 
```

```python
sp500_tickers = table[0]['Symbol'].tolist() # seleccionamos los tickers y los guardamos en una lista 
#sp500_tickers
```

```python
data=[]
```

```python
for i in sp500_tickers: #descargamos todos los datos de las compañias con yahoo finance y los guardamos en un Dataframe 
    stock=yf.Ticker(i)
    ratios=stock.info
    ratios['Ticker']=i
    
    data.append(ratios)
df=pd.DataFrame(data)
```

```python
cols = ['Ticker'] + [col for col in df.columns if col != 'Ticker'] #mostramos los Tickers de las compañias como primera columna 
df = df[cols]
df
```

```python
columnas=df.columns.tolist()
columnas #muestra todas las columnas del dataframe , seleccionaremos las que nos interesan más adelante 
```

```python
df.sector.value_counts() # muestra los  diferentes sectores 
```

```python
print(df[['dividendYield', 'priceToBook', 'trailingPE', 'profitMargins', 'returnOnEquity', 'debtToEquity', 'revenueGrowth']].isnull().sum()) #cuantos valores nulos hay en las columnas que nos interesan
```

```python
columns_of_interest = ['trailingPE', 'forwardPE', 'priceToBook','dividendYield', 'returnOnEquity', 'profitMargins','returnOnAssets','debtToEquity']

df[columns_of_interest] = df[columns_of_interest].apply(pd.to_numeric, errors='coerce') #para evitar errores de formato 

#calculamos la media por sectores para tener una referencia 
sector_mean = df.groupby('sector')[columns_of_interest].mean()


print(sector_mean)
```

## Seleccion de Filtros para encontra compañias subvaluadas

```python
df[df['dividendYield']>0.10] #Compañias con dividendos mayores a 10% , esto no quiere decir que sea buena inversion, hay mas factores a tener en cuenta 
```

```python
df[(df['dividendYield']>0.05)&(df['priceToBook']<1)] #compañias con dividendo de 5% y una relacion valor de mercado-valor contable baja 
```

```python
#filtros que indican posibles compañias que estan subvaluadas, estos filtros se pueden modificar a medida 
df_subvaluadas = df[
    (df['dividendYield'] > 0.01) &   # Dividendos > 1%
    (df['priceToBook'] < 10) &       # Relación Precio-valor contable < 10
    (df['trailingPE'] < 35) &       # Relación Precio- Beneficio  < 35
    (df['profitMargins'] > 0.03) &       # Margen de utilidad > 3%
    (df['returnOnEquity'] > 0.02) &    # ROE > 2%
    (df['debtToEquity'] < 100) &           # Deuda/patrimonio < 100
    (df['revenueGrowth'] >= 0)   # Crecimiento de ingresos  positivo o nulo
]
```

```python
df_subvaluadas # Una vez realizado el filtro podemos centrar nuestro analisis value investing en estas compañias 
```
